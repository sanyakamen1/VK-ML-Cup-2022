```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import SGDClassifier
```

```python
X_test = pd.read_csv('test.csv')
```

```python
sample_submission = pd.read_csv('sample_submission.csv')
```

```python
train_file = pd.read_csv('train.csv')
```

```python
target = train_file['category']
```

```python
data = train_file.drop('category', axis=1)
```

```python
X_test['text'] = X_test['text'].str.lower()
```

```python
data['text'] = data['text'].str.lower()
```

```python
cv = CountVectorizer()
cv.fit(data['text'])

len(cv.vocabulary_)
```

```python
data_transform = cv.transform(data['text'])
```

```python
test_transform = cv.transform(X_test['text'])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_val, y_train, y_val = train_test_split(data_transform, target, test_size=0.30, random_state=0)
```

```python
sgd = SGDClassifier(loss='log_loss', random_state=0, verbose=1, n_jobs=3, max_iter=100000)
```

```python
sgd.fit(X_train, y_train)
```

```python
sgd.score(X_train, y_train) # Train score
```

```python
sgd.score(X_val, y_val) # Val score
```

```python
pred_probs1 = pd.DataFrame(sgd.predict_proba(test_transform))
```

```python
pred_probs1.head()
```

```python
prob_values1 = pd.DataFrame([max(i) for i in pred_probs1.values])
```

```python
prob_values1.head()
```

```python
pred1 = sgd.predict(test_transform)
```

```python
sub = X_test.copy()
```

```python
sub.head()
```

```python
sub = sub.drop('text', axis=1)
```

```python
sub['category'] = pred1
```

```python
sub['prob'] = prob_values1
```

```python
sub.head()
```

```python
sub = sub[sub.prob >= 0.7]
```

```python
sub.head()
```

```python
sub = sub.drop('prob', axis=1)
```

```python
sub.head()
```

```python
sub.to_csv('sgd_07.csv')
```
